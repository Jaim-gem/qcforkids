# **How to become a Quantum Scientist**
&nbsp; 

There are *many jobs* that use Quantum Science and many large companies around the world that hire Quantum Scientists.<br>

Quantum Science is a **very big** area and there are so many things that you can do; *Quantum Computing is just one of 
them!*
&nbsp; 

First things first, you need to study **Math** and **Science**.<br>

In Secondary school, the two main Science subjects that you should take are **Physics** and **Chemistry**. *Also*, never forget that Math is *your best friend*.  

At University, you can either take the path of *Pure Science* like Physics, Math, Chemistry and Computer Science **or** you can do *Applied Science* like **Electrical Engineering** if you are more interested in *building* quantum computers.


If you are interested particularly in *Quantum Computing*, you can become a: 

- **Quantum Hardware Engineer**: This is a person who works on the actual *designing and building* of quantum computers. 


- **Quantum Software Engineer**: This is a person who writes *code* to be able to do calculations on quantum computers. 


- **Quantum Computing Scientist**: This is a person who *investigates* in Quantum computing. They can be involved in both the *building* of the quantum computers as well as *writing code* for calculations. 


- **Quantum Computing Applications Scientist**: This is a person who looks at the many things that Quantum Computing can be used for. They *investigate* and *write code* for a specific use for Quantum Computing.
